# Analysing PSFs

```python
from astropy.io import fits
from matplotlib import pyplot
import aotools
```

```python
uncorrected_psf = fits.getdata('uncorrected_psf.fits')
tip_tilt_psf = fits.getdata('tip_tilt_psf.fits')
corrected_psf = fits.getdata('corrected_psf.fits')
```

```python
fig, axs = pyplot.subplots(1, 3)
ax1, ax2, ax3 = axs
im1 = ax1.imshow(uncorrected_psf)
im2 = ax2.imshow(tip_tilt_psf)
im3 = ax3.imshow(corrected_psf)
fig.tight_layout()
```

```python
fig, axs = pyplot.subplots(1, 3)
ax1, ax2, ax3 = axs
im1 = ax1.imshow(uncorrected_psf)
fig.colorbar(im1)
im2 = ax2.imshow(tip_tilt_psf)
fig.colorbar(im2)
im3 = ax3.imshow(corrected_psf)
fig.colorbar(im3)
fig.tight_layout()
```

```python
fig, axs = pyplot.subplots(1, 3)
ax1, ax2, ax3 = axs
im1 = ax1.imshow(uncorrected_psf, vmin=0, vmax=corrected_psf.max())
im2 = ax2.imshow(tip_tilt_psf, vmin=0, vmax=corrected_psf.max())
im3 = ax3.imshow(corrected_psf, vmin=0, vmax=corrected_psf.max())
fig.tight_layout()
```

```python
uncorrected_cut = aotools.azimuthal_average(uncorrected_psf)
tip_tilt_cut = aotools.azimuthal_average(tip_tilt_psf)
corrected_cut = aotools.azimuthal_average(corrected_psf)
```

```python
pyplot.plot(uncorrected_cut, label='uncorrected PSF')
pyplot.plot(tip_tilt_cut, label='Tip/Tilt corrected PSF')
pyplot.plot(corrected_cut, label='AO corrected PSF')
pyplot.legend()
pyplot.yscale('log')
```

```python
pyramid_psf = fits.getdata('pyramid_psf.fits')
pyplot.imshow(pyramid_psf)
pyplot.colorbar()
```

```python
pyramid_cut = aotools.azimuthal_average(pyramid_psf)
```

```python
pyplot.plot(uncorrected_cut, label='uncorrected PSF')
pyplot.plot(tip_tilt_cut, label='Tip/Tilt corrected PSF')
pyplot.plot(corrected_cut, label='AO corrected PSF')
pyplot.plot(pyramid_cut, label='Pyramid WFS PSF')
pyplot.legend()
pyplot.yscale('log')
```
